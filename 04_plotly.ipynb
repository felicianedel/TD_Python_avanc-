# Plotly

## Exemples de graphique

```python
import plotly.graph_objects as go 
fig = go.Figure(
    data=[
        go.Bar(x=['A','B','C'], y=[1,2,3])
    ],
    layout=go.Layout(
        title = go.layout.Title(text="A Figure Specified By A Graph Object")
    )
)
fig.show()
```

```python
dict_of_fig = dict({
    "data":[{
        "type":"bar",
        "x":["A","B","C"],
        "y":[1,3,2]
    }],
    "layout":{
        "title":{"text": "A Figure Specified By A Graph Object With A Dictionary"}
    }
})

fig = go.Figure(dict_of_fig)
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2)

fig.add_trace(go.Bar(y=[2,3,5], marker_color="green"), row=1, col=1)
fig.add_trace(go.Scatter(y=[8,3,2,4], mode="lines"), row=1, col=2)

fig.update_layout(title_text="Subplots with 1 row and 2 columns", height=400)

fig.show()
```

## Scatter plot

```python
import plotly.express as px

df = px.data.iris()
print(df.head())
fig = px.scatter(df, x="sepal_width", y = "sepal_length", color="species")
fig.show()
```

## Scatter matrix

```python
import plotly.express as px
df = px.data.iris()
fig=px.scatter_matrix(df,
                      dimensions=["sepal_width","sepal_length","petal_width","petal_length"],
                      color="species")

fig.show()
```

## Scatter bubble

```python
import plotly.express as px
df = px.data.gapminder()
df = df.query("year==2007")

fig = px.scatter(df,
                 x = "gdpPercap",
                 y="lifeExp", size="pop",
                 color="continent",
                 hover_name="country",
                 log_x=True,
                 size_max=60)

fig.show()
```

## Scatter bubble series

```python
import plotly.express as px
df = px.data.gapminder()
fig=px.scatter(df,
                 x = "gdpPercap",
                 y="lifeExp", 
                 size="pop",
                 color="continent",
                 hover_name="country",
                 log_x=True,
                 size_max=45,
                 
                 #facet_col="continent",
                 animation_frame="year",
                 animation_group="country",
                 range_x=[100,100000],
                 range_y=[25,90])

fig.update_layout(height=500, width=1000, title_text="Gapminder")
fig.show()
```

## Histogram

```python
import plotly.express as px
df = px.data.tips()
print(df.head())
fig = px.histogram(df, x="total_bill", y="tip", color="sex", marginal="rug")
fig.show()
```

## Bar chart

```python
import plotly.express as px
df = px.data.tips()
fig= px.bar(df, x="sex", y="total_bill", color="smoker", barmode="group")
fig.show()
```

## Bar with facet

```python
import plotly.express as px
df = px.data.tips()

fig = px.bar(df,
             x="sex",
             y="total_bill",
             color="smoker",
             barmode="group",
             
             facet_row="time",
             facet_col="day",
             category_orders={
                 "day": ["Thur", "Fri", "Sat", "Sun"],
                 "time":['Lunch', "Dinner"]
             }
             )

fig.show()
```

```python
import plotly.express as px
df = px.data.wind()
print(df.head())

fig = px.bar_polar(df,
                   r="frequency",
                   theta="direction",
                   color="strength",
                   template="plotly_dark",
                   color_discrete_sequence= px.colors.sequential.Plasma_r)

fig.show()
```

```python
import plotly.express as px

df= px.data.gapminder().query("year == 2007")
fig = px.sunburst(df, path=['continent','country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'])

fig.show()
```

## Box plot

```python
import plotly.express as px
df = px.data.tips()
fig=px.box(df, x="day", y="total_bill",color="smoker", notched=True)

fig.show()
```

## Histogram 2D

```python
import plotly.express as px

df = px.data.iris()
print(df.head())

fig = px.density_heatmap(df,
                         x="sepal_width",
                         y="sepal_length",
                         marginal_x="rug",
                         marginal_y="histogram")
fig.update_layout(width=800, height=400, margin=dict(l=10, r=10, t=20, b=10))
fig.show()
```

## Choropleth

```python
import plotly.express as px
df = px.data.gapminder()

fig = px.choropleth(df,
                    locations="iso_alpha",
                    color="lifeExp",
                    hover_name="country",
                    animation_frame="year",
                    range_color=[20,80])
fig.update_layout(height=800)
fig.show()
```

```python
import plotly.express as px
df = px.data.carshare()

fig = px.scatter_mapbox(df,
                    lat="centroid_lat",
                    lon="centroid_lon",
                    color="peak_hour",
                    size="car_hours",
                    color_continuous_scale=px.colors.cyclical.IceFire,
                    size_max=15, zoom=10,
                    mapbox_style="carto-positron")
fig.show()
```

## Indicators

```python
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(
    go.Indicator(value = 200,
                 delta={'reference':160},
                 gauge={'axis':{'visible':False}},
                 domain= {'row':0, 'column':0})
)

fig.add_trace(
    go.Indicator(value=120,
                 gauge= {'shape':"bullet",'axis': {'visible':False}},
                 domain= {'x':[0.05,0.5], 'y':[0.15,0.35]})
)
fig.add_trace(
    go.Indicator(value=300,
                 mode="number+delta",
                 domain= {'row':0, 'column':1})
)

fig.add_trace(go.Indicator(mode="delta", value=40, domain={'row':1, 'column':1}))

fig.update_layout(
    grid={'rows':2, 'columns':2, 'pattern':"independent"},
    template= {'data': {'indicator' : [{
        'title':{'text':"Speed"},
        'mode':"number+delta+gauge",
        'delta':{'reference':10}
    }]}}
)
fig.show()
```
