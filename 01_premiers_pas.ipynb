```python
# ========== Importation des librairies ========= #
import dash
from dash import html, dcc, Input, Output, callback

# ======== Initialisation de l'application ====== #
app = dash.Dash()

# ============== Layout ========================= #
app.layout = html.Div(children = [ 
    html.H1(children = "Hello Dash!"),

    dcc.Input(id='input-1', value='', type='text'),
    html.Div(id='div-1',  children='')
])

# ================ Callbacks ==================== # 
# ajouter l'interactivité dans l'application
@callback(
    Output(component_id='div-1', component_property='children'), #output
    [Input(component_id='input-1', component_property='value')]  #input
)

def  update_div(input):
    return f'Texte introduit : {input}'

# ================= Run server ================== #
# Démarrer l'application Dash
if __name__ == '__main__':
    app.run(debug=True, port=8050, jupyter_mode="external")
```
